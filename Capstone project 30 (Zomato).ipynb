Check rating category with the price range of the restaurant w.r.t Value Counts function by applying group by function

```python
df.approx_cost = pd.to_numeric(df.approx_cost,errors='coerce')
```

```python
df.approx_cost = df.approx_cost.fillna(0)
```

```python
df.approx_cost = df.approx_cost.astype(int)
```

```python
df["Price_range(2)"] = pd.cut(x = df["approx_cost"], bins = [0,500,1000,1500,2000,2500,3000,4000,9000],
                                   labels=["cost < 500", "500-1000", "1000-1500", "1500-2000", "2000-3000", "3000-4000", "4000-5000",
                                             "5000 < cost "])
```

```python
df.groupby(['rate'])['Price_range(2)'].value_counts()
```
